## 2nd min edit dist

```python
def edit_distance(s, t):
    m, n = len(s), len(t)

    dp = [[0]*(n+1) for _ in range(m+1)]

    # initialize
    for i in range(m+1):
        dp[i][0] = i
    for j in range(n+1):
        dp[0][j] = j

    # fill table
    for i in range(1, m+1):
        for j in range(1, n+1):
            if s[i-1] == t[j-1]:
                dp[i][j] = dp[i-1][j-1]
            else:
                dp[i][j] = 1 + min(
                    dp[i-1][j],     # delete
                    dp[i][j-1],     # insert
                    dp[i-1][j-1]    # replace
                )

    return dp[m][n]


# User input
word1 = input("Enter wrong word: ")
word2 = input("Enter correct word: ")

# Calculate distance
dist = edit_distance(word1, word2)

# Output
print("\nWord Entered :", word1)
print("Correct Word :", word2)
print("Minimum Edit Distance :", dist)
```

### 3rd vertbi

```python
# small corpus
corpus = [("I","PRON"),("eat","VERB"),("rice","NOUN")]

# probabilities (hardcoded for simplicity)
transition = {
    "<s>":{"PRON":1},
    "PRON":{"VERB":1},
    "VERB":{"NOUN":1}
}

emission = {
    "PRON":{"I":1},
    "VERB":{"eat":1},
    "NOUN":{"rice":1}
}

tags = ["PRON","VERB","NOUN"]

def viterbi(words):
    V = [{}]
    path = {}

    # init
    for t in tags:
        V[0][t] = transition.get("<s>",{}).get(t,0) * emission.get(t,{}).get(words[0],0)
        path[t] = [t]

    # loop
    for i in range(1,len(words)):
        V.append({})
        newpath = {}

        for t in tags:
            max_prob, prev = max(
                (V[i-1][pt] * transition.get(pt,{}).get(t,0) *
                 emission.get(t,{}).get(words[i],0), pt)
                for pt in tags
            )
            V[i][t] = max_prob
            newpath[t] = path[prev] + [t]

        path = newpath

    # result
    _, final_tag = max((V[-1][t], t) for t in tags)
    return path[final_tag]


# input
words = input("Enter sentence: ").split()

result = viterbi(words)

print("\nPOS Tags:")
for w,t in zip(words,result):
    print(w,"→",t)
```
