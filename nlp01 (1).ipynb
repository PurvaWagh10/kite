#### Title : Text Preprocessing - Implement text preprocessing techniques, including tokenization and normalization, and apply it to any text data.

```python
import re
```

```python
import nltk
```

```python
nltk.download('punkt')
nltk.download('stopword')
nltk.download('wordnet')
```

```python
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
```

```python
text="In 2025, there were 25 students in the class. Each student studied for 3 hours a day and completed 2 assignments every week. The teacher gave 10 points for each assignment."
print("original test:", text)
```

```python
text=text.upper()
print("upper case conversion:", text)
```

```python
text = re.sub(r'\d+', '', text)
text
```

```python
text = re.sub(r'[^A-Z\s]', '', text)
print("After Removing Numbers & Punctuation:", text)
```

```python
tokens = word_tokenize(text)
print("After Tokenization:", tokens)
```

```python
stop_words = set(stopwords.words('english'))
tokens = [word for word in tokens if word.lower() not in stop_words]
print("After Stopword Removal:", tokens)
```

```python
lemmatizer = WordNetLemmatizer()
lemmatized_tokens = [lemmatizer.lemmatize(word.lower()) for word in tokens]
print("After Lemmatization:", lemmatized_tokens)
```

```python
stemmer = PorterStemmer()
stemmed_tokens = [stemmer.stem(word) for word in tokens]
print("After Stemming:", stemmed_tokens)
```
