```python
import nltk
from nltk.wsd import lesk
from nltk.corpus import wordnet as wn
from nltk.tokenize import word_tokenize
```

```python
import nltk
nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
word = "bank"
```

```python
sent1 = "I went to the bank to deposit money"
sent2 = "The river overflowed the bank after heavy rain"
```

```python
tokens1 = word_tokenize(sent1)
tokens2 = word_tokenize(sent2)
```

```python
sense1 = lesk(tokens1, word)
sense2 = lesk(tokens2, word)
```

```python
print("Sentence 1:", sent1)
print("Sense:", sense1)
print("Definition:", sense1.definition())
```

```python
print("\nSentence 2:", sent2)
print("Sense:", sense2)
print("Definition:", sense2.definition())
```

```python
print("\nAll Synsets of 'bank':")
for syn in wn.synsets("bank"):
    print(syn.name(), ":", syn.definition())
```
